```python
import pandas as pd
import numpy as np

import json

import matplotlib
import matplotlib.pyplot as plt
from matplotlib import cm

matplotlib.rcParams['figure.figsize'] = (10, 10)
```

```python
file_name = 'USvideos.csv' 
my_df = pd.read_csv(file_name)
my_df.head()
```

```python
print('NaN CELLS {}:\n{}\n'.format('BEFORE DIRTYING', my_df.isna().sum()))
```

```python
for i in range(1, 36000, 30):
    my_df.loc[[i], my_df.columns[4]] = np.nan
    my_df.loc[[i], my_df.columns[10]] = np.nan
```

```python
print('NaN CELLS {}:\n{}\n'.format('AFTER DIRTYING', my_df.isna().sum()))
```

```python
avg = my_df[my_df.columns[10]].mean()
for i in range(1, 36000, 30):
    my_df.loc[[i], my_df.columns[10]] = avg
my_df.fillna(method='ffill', inplace=True)
```

```python
print('NaN CELLS {}:\n{}\n'.format('AFTER CLEANING', my_df.isna().sum()))
```

```python
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from scipy import stats

import datetime
import json

```

```python
type_int_list = ['likes', 'dislikes', 'comment_count']
for column in type_int_list:
    my_df[column] = my_df[column].astype(int)

type_str_list = ['category_id']
for column in type_str_list:
    my_df[column] = my_df[column].astype(str)
```

```python
print('Null hypothesis: Trending videos receive an average of over 3.5k dislikes each')
print('Alternate hypothesis: Trending videos receive less than or equal to  3.5k dislikes each')
alpha = 0.05
x =my_df['dislikes'].mean()
mu = 3500
std = my_df['dislikes'].std()
root_n =my_df['dislikes'].count() ** (1/2)
z = (x-mu)/(std/root_n)
print('alpha: {}\nmu: {}\nx: {}\nstd: {}\nroot_n: {}\nz: {}'.format(alpha, mu, x, std, root_n, z))
print('Putting this value in the z-table we get:-')
print('p value:', 0.96)
print('Therefore, we reject the null hypothesis.')
```

```python
id_to_category = {}

with open('US_category_id.json', 'r') as f:
    data = json.load(f)
    for category in data['items']:
        id_to_category[category['id']] = category['snippet']['title']

id_to_category
```

```python
my_df.insert(4, 'category', my_df['category_id'].map(id_to_category))
my_df[['category_id', 'category']].head()
```

```python
 
top10_views = my_df[['views', 'title', 'likes', 'dislikes', 'comment_count', 'publish_time', 'trending_date', 'category_id', 'category']].sort_values(ascending=False, by=['likes'])
top10_views = top10_views.drop_duplicates(subset="title", keep='first')
top10_views = top10_views.head(10)
top10_views[['likes', 'title']].head(10)
# Label list
label = list(top10_views['title'])

# Plot data
ax = top10_views.plot.bar(x='title', y='likes', rot=85, edgecolor='black',
                         linewidth=2, align='center', figsize=(8,4), width=0.8, legend=None)

# Set axis labels / title / legend
ax.set_ylabel("Likes", size=12)
ax.set_xlabel("Titles", size=12)
ax.set_title("Top 10 Most Liked Videos", size=15)

plt.show()
```

```python
def visualize_statistics(my_df, id_list): # taking a list of video ids
    target_df = my_df.loc[id_list]
    
    ax = target_df[['views', 'likes', 'dislikes', 'comment_count']].plot.bar()
    
    # customizes the video titles, for asthetic purposes for the bar chart
    labels = []
    for item in target_df['title']:
        labels.append(item[:10] + '...')
    ax.set_xticklabels(labels, rotation=45, fontsize=10)
    
    plt.show()

def visualize_like_dislike(my_df, id_list):
    target_df = my_df.loc[id_list]
    
    ax = target_df[['likes', 'dislikes']].plot.bar(stacked=True)
    
    # customizes the video titles, for asthetic purposes for the bar chart
    labels = []
    for item in target_df['title']:
        labels.append(item[:10] + '...')
    ax.set_xticklabels(labels, rotation=45, fontsize=10)
    
    plt.show()
sample_id_list = my_df.sample(n=10, random_state=4).index # creates a random sample of 10 video IDs
sample_id_list
```

```python
visualize_statistics(my_df, sample_id_list)
```

```python
visualize_like_dislike(my_df, sample_id_list)
```
